```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
pic = np.ones((500,500,3), dtype='uint8')
pic = cv.Sobel(pic, cv.CV_64F, 1, 1, ksize=3)
print(pic.sum())

plt.imshow(pic)
plt.show()
```

```python
# decomposed separable filter
filter_vertical = np.array([1, 2, 1])
filter_horitzontal = np.array([-1, 0, 1])

# image
part = np.array([
    [115, 65, 120, 103, 101],
    [124, 60, 100, 89, 110],
    [170, 67, 114, 87, 75]
    ], dtype=np.uint8)

# Apply this filter to an image (to 3 consecutive pixels with stride 1 and no padding). Find max(𝑟)−min(𝑟), difference between maximum and minimum values of a filtered image
filtered_vertical = np.zeros((1, 5))
for i in range(5):
    filtered_vertical[0, i] = np.sum(part[:, i] * filter_vertical)

filtered_image = np.zeros((1, 3))
for i in range(3):
    filtered_image[0, i] = np.sum(filtered_vertical[:, i:i+3] * filter_horitzontal)

print(filtered_image.max() - filtered_image.min())

plt.imshow(filtered_image)
plt.show()
```

```python
part = np.array([
    [115, 65, 120, 103, 101],
    [124, 60, 100, 89, 110],
    [170, 67, 114, 87, 75]
    ], dtype=np.uint8)

# apply sobel filter - only horizontal
part = cv.Sobel(part, cv.CV_64F, 1, 0, ksize=3)

plt.imshow(part)
plt.show()

```
